```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
btc = pd.read_html('https://coinmarketcap.com/currencies/bitcoin/historical-data/?start=20130428&end=20190906')
btc = btc[0]
dates = btc['Date']
dates = pd.to_datetime(dates)
btc.drop(labels='Date',axis=1,inplace=True)
btc['Date'] = dates
btc = btc.sort_values(by='Date',axis=0,ascending=True)
btc.reset_index(inplace=True)
btc.drop(labels='index',axis=1,inplace=True)
btc.set_index(keys='Date',inplace=True)
```

```python
btc['percentage_daily_change'] = (btc['Close**'] / btc['Close**'].shift(1)) - 1 
```

```python
btc.head()
```

```python
btc['percentage_daily_change'].plot(figsize=(8,5))
```

```python
btc['percentage_daily_change'].loc['2019-01-01':'2019-09-05'].plot(figsize=(8,5))
```

```python
btc['percentage_daily_change'].loc['2015-10-01':'2018-01-01'].plot(figsize=(8,5))
```

```python
avg_returns_all_time = btc['percentage_daily_change'].mean()
avg_returns_all_time
```

```python
avg_returns_annual = btc['percentage_daily_change'].mean() * 365
avg_returns_annual
```

```python
print(str(round(avg_returns_annual,5)*100) + ' %')
```

```python
avg_returns_2015_2018 = btc['percentage_daily_change'].loc['2015-10-01':'2018-01-01'].mean()
avg_returns_2015_2018
```

```python
btc['Log Returns'] = np.log(btc['Close**'] / btc['Close**'].shift(1))
```

```python
btc['Log Returns'].plot(figsize = (8,5))
```

```python
log_return_d = btc['Log Returns'].mean()
log_return_d
```

```python
log_return_a = btc['Log Returns'].mean() * 365
log_return_a
```

```python
print(str(round(log_return_a,4)*100) + ' %')
```
