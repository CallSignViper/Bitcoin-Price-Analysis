Granger Causlity Test

```python
import pandas as pd
```

```python
dow_jones = pd.read_excel('DJI.xlsx',index_col='Effective date ',parse_dates=True,)
```

```python
dow_jones.dropna(axis=0,inplace=True)
```

```python
dow_jones.sort_index(ascending=False,inplace=True)
```

```python
from statsmodels.tsa.stattools import grangercausalitytests
```

```python
btc = pd.read_html('https://coinmarketcap.com/currencies/bitcoin/historical-data/?start=20130428&end=20200303')
btc = btc[2]
#btc.sort_values(by='Date',ascending=True,inplace=True)
btc.reset_index(inplace=True)
btc.drop(labels='index',inplace=True,axis=1)
dates = btc['Date']
dates = pd.to_datetime(dates)
btc.drop(labels='Date',inplace=True,axis=1)
btc['Date'] = dates
btc.set_index(keys=dates,inplace=True)
btc.drop(labels='Date',inplace=True,axis=1)
```

```python
date_index = pd.date_range(start='2013-04-28', end='2020-02-29',freq='D')
```

```python
btc_ = btc.loc["2020-02-29":"2013-04-28"]
```

```python
btc_.drop(labels=['High', 'Low', 'Market Cap', 'Open*', 'Volume'],axis=1,inplace=True)
```

```python
dow_jones_ = dow_jones.loc["2020-02-29":"2013-04-28"]
```

```python
dow_jones_.index = pd.to_datetime(dow_jones_.index)
```

```python
btc_.drop(btc_.index[0],inplace=True)
```

```python
dataset = dow_jones_.join(btc_)
```

```python
dataset.head()
```

```python
help(grangercausalitytests)
```

```python
grangercausalitytests(dataset[['Close**','Dow Jones Industrial Average']],maxlag=60)
```
